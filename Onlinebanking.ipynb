```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("D:\\Github\\Aktien\\Input\\Umsaetze.csv", sep = ";", decimal=',')
df.tail(2)
```

```python
df = df.filter(items=['Valutadatum', 'Saldo nach Buchung'])
df
```

```python
df['Jahr'] = df.Valutadatum.str.slice(-4)
df['Monat'] = df.Valutadatum.str.slice(-7,-5)
df['Tag'] = df.Valutadatum.str.slice(-10,-8)
df.dtypes
```

```python
df
```

```python
df = df.sort_values(
    by=['Saldo nach Buchung'],
    ascending = False,
    ignore_index = True)
df
```

```python
df = df.sort_values(
    by=['Jahr', 'Monat', 'Tag'],
    ascending = False,
    ignore_index = True)
df
```

```python


# plt.figure(figsize=(22, 5))
plt.style.use('seaborn')
plt.grid(True)

plt.plot(df['Saldo nach Buchung'])
plt.show()
```
